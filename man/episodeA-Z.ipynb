```python
def factorize(n, m=2):
    if n == 1:
        return tuple()
    elif n % m == 0:
        return (m,) + factorize(n//m)
    elif n < m**2:
        return (n,)
    else:
        return factorize(n, 3 if m == 2 else m+2)

assert (2, 2, 3) == factorize(12) and (2, 3, 7) == factorize(42)
assert (3, 3, 5) == factorize(45) and (3, 5, 5) == factorize(75)
```

```python
from math import *
from random import *

def is_prime(n):
    if n in {2, 3}:
        return True
    if n == 1 or n % 2 == 0:
        return False
    for d in range(3, 1+int(sqrt(n)), 2):
        if n % d == 0:
            return False
    else:
        return True

assert [2, 3, 5, 7, 11] == [e for e in range(12) if is_prime(e)]

def prime(n, certainty=5): # returns when the prob. exceeds 1 - 0.5 ** certainty.
    if n < 4:
        return max(2, n) # prime p >= 2
    if n % 2 == 0:
        n += 1
    for p in range(n, 2*n): # prime p exists where n < p < 2n - 2 when n > 1.
        for _ in range(certainty):
            a = 2 + randrange(p-3) # 2 <= a <= p-2; 2+0 <= a <= 2+p-4.
            if 1 != a ** (p-1) % p: # rabin miller's primality test.
                break
        else:
            return p

assert [2, 2, 2, 3, 5, 5, 7, 7, 11, 11] == [prime(n) for n in range(0, 10)]
```

```python
def count_bst(n, memos={}): # binary search tree
    if n == 0:
        return 1
    if n not in memos:
        memos[n] = sum(count_bst(k) * count_bst(n-k-1) for k in range(n))
    return memos[n]

assert [1, 1, 2, 5, 14] == [count_bst(e) for e in (0, 1, 2, 3, 4)]
```

```python
ops = {'*': 1, '/': 1, '+': 2, '-': 2, '(': 3}

def postfix(a):
    q, s = [], []
    for e in a:
        if e == '(':
            s.append(e)
        elif e == ')':
            while s[-1] != '(':
                q.append(s.pop())
            s.pop()
        elif e not in ops:
            q.append(e)
        else:
            while s and ops[s[-1]] <= ops[e]:
                q.append(s.pop())
            s.append(e)
    while s:
        q.append(s.pop())
    return q

def evaluate(q): # rpn: reverse polish notation.
    s = []
    for e in q:
        if e in ops:
            e = eval("{2}{1}{0}".format(s.pop(), e, s.pop()))
        s.append(e)
    return s.pop()

assert '234*+23+4*+' == ''.join(postfix(list('2+3*4+(2+3)*4')))
assert 34 == evaluate(postfix(list('2+3*4+(2+3)*4')))
```

```python
inf = 2**31 - 1 # or sys.maxvalue

def optimal_tour(g): # TSP http://www.youtube.com/watch?v=aQB_Y9D5pdw
    def popped(s, i):
        l = list(s)
        l.pop(i)
        return tuple(l)
    # g(v, set) is min([g[v, e] + g(e, set-{e}) for e in set])
    def mapped(v, s, memos={}):
        def computed():
            if s:
                return min(
                    ((w, g[v][w] + mapped(w, popped(s, i))[1])
                     for i, w in enumerate(s)
                     if g[v][w] != inf),
                    key=lambda e: e[1])
            else:
                return (None, inf) if g[v][0] is inf else (0, g[v][0])
        k = (v, s) 
        if k not in memos:
            memos[k] = computed()
        return memos[k]
    return mapped(0, tuple(range(1, len(g))))[1]

g = [None] * 4
g[0] = [0, 10, 15, 20]
g[1] = [5, 0, 9, 10]
g[2] = [6, 13, 0, 12]
g[3] = [8, 8, 9, 0]

assert 35 == optimal_tour(g)

def floyd_warshal(g):
    d = [r[:] for r in g]
    n = len(g) # graph in n x n matrix
    for k in range(n):
        for i in range(n):
            for j in range(n):
                d[i][j] = min(d[i][j], d[i][k] + d[k][j])
    return d

floyd_warshal(g)
```

```python
# Given a list of intervals [start, end], collapse intervals that overlap.
def collapsed(intervals):
    intervals = sorted(intervals, key=lambda e: e[0]) # sorted by start.
    collapsed = []
    tail = 0
    for head in range(1, 1 + len(intervals)):
        if head == len(intervals) or intervals[head - 1][1] < intervals[head][0]:
            collapsed.append((intervals[tail][0], intervals[head-1][1]))
            tail = head
    return collapsed

assert [(2, 8), (10, 12)] == collapsed([(2, 6), (4, 8), (10, 12)])
assert [(2, 4), (6, 12)] == collapsed([(2, 4), (6, 10), (8, 12)])
```

```python
def minmax(a, start=0, stop=None):
    if stop is None:
        stop = len(a)
    if stop - start < 2:
        return (a[start], a[start])
    else:
        mid = (start + stop - 1)//2
        l = minmax(a, start, mid+1)
        r = minmax(a, mid+1, stop)
        return (min(l[0], r[0]), max(l[1], r[1]))

assert (1, 9) == minmax([1, 3, 5, 7, 9, 2, 4, 6, 8])

def peak(a, start=0, stop=None):
    if stop is None:
        stop = len(a)
    mid = (start + stop - 1)//2
    if a[mid-1] < a[mid] > a[mid+1]:
        return mid
    elif a[mid-1] < a[mid] < a[mid+1]:
        return peak(a, mid+1, stop)
    else:
        return peak(a, start, mid)

assert 6 == peak([1, 5, 7, 9, 15, 18, 21, 19, 14])
```

```python
def three_sum_closest(a, q=0):
    a, n = sorted(a), len(a)
    tuples = []
    for i in range(n-2):
        j, k = i+1, n-1
        while j < k:
            diff = abs(q - sum((a[i], a[j], a[k])))
            if not tuples or mindiff > diff:
                mindiff = diff
                tuples = [(a[i], a[j], a[k])]
            elif mindiff == diff:
                tuples.append((a[i], a[j], a[k]))
            cmp = q - sum((a[i], a[j], a[k]))
            if cmp < 0:
                j += 1
            elif cmp > 0:
                k -= 1
            else:
                j += 1
                k -= 1
    return tuples

def three_sum(a, q=0):
    a, n = sorted(a), len(a)
    tuples = []
    for i in range(n-3):
        j, k = i+1, n-1
        while j < k:
            cmp = q - sum((a[i], a[j], a[k]))
            if cmp < 0:
                j += 1
            elif cmp > 0:
                k -= 1
            else:
                tuples.append((a[i], a[j], a[k]))
                j += 1
                k -= 1
    return tuples

assert [(-1, 1, 2)] == three_sum_closest([-1, 2, 1, -4])
assert [(-1, -1, 2), (-1, 0, 1)] == three_sum([-1, 0, 1, 2, -1, -4], 0)
```

```python
def next_permutate(a):
    n = len(a)
    i = next((i for i in range(n-2, -1, -1) if a[i] < a[i+1]), -1)
    if i >= 0:
        j = next(j for j in range(n-1, i, -1) if a[i] < a[j])
        a[i], a[j] = a[j], a[i]
        a[i+1:] = a[-1:i:-1]
        return a

assert next_permutate([4, 3, 2, 1]) is None
assert [2, 1, 3, 4] == next_permutate([1, 4, 3, 2])
```

```python
class BNode:
    def __init__(self, value=None, left=None, right=None, parent=None):
        self.value, self.left, self.right, self.parent = value, left, right, parent

# http://www.youtube.com/watch?v=PAYG5WEC1Gs&feature=plcp
def parse(preorder, inorder):
    if preorder:
        v = preorder[0]
        k = inorder.index(v)
        return BNode(v,
            parse(preorder[1:k+1], inorder[:k]),
            parse(preorder[k+1:], inorder[k+1:]))

# preorder: abcdefg
# inorder:  cdebagf
# tree:      a
#         b    f
#       c     g
#        d
#         e

tree = parse("abcdefg", "cdebagf")
assert ["a", "b", "c", "d", "e", "f", "g"] == [e.value for e in (
    tree, tree.left, tree.left.left, tree.left.left.right,
    tree.left.left.right.right, tree.right, tree.right.left)]
assert all((e is None for e in (
    tree.left.right, tree.left.left.left, tree.left.left.right.left,
    tree.left.left.right.right.left, tree.left.left.right.right.right,
    tree.right.right, tree.right.left.left, tree.right.left.right)))
```

```python
def diameter(tree): # return (d, h)
    if tree is None:
        return (0, 0)
    l = diameter(tree.left)
    r = diameter(tree.right)
    d = max(l[0], r[0], 1 + l[1] + r[1])
    h = 1 + max(l[1], r[1])
    return (d, h)

# tree input:   a
#             b
#          c    f
#           d     g
#            e
tree = parse('abcdefg', 'cdebfga')
assert (6, 5) == diameter(tree)
```

```python
def backtrack(candidate, expand_out, reduce_off):
    if not reduce_off(candidate):
        for e in expand_out(candidate):
            candidate.append(e)
            backtrack(candidate, expand_out, reduce_off)
            candidate.pop()

def color_vertex(g):
    def expand_out(a):
        v = len(a) # vertex v
        return [c for c in set(a) if peaceful_at(a, v, c)] + [max(a)+1]

    def peaceful_at(a, v, c):
        return all([g[v][w] == 0 or a[w] != c for w in range(v)])

    def reduce_off(a):
        if reduce_off.reduced or len(a) == len(g):
            reduce_off.reduced.append((max(a)+1, a[:]))
            return True

    reduce_off.reduced = []
    backtrack([0], expand_out, reduce_off)
    return min(reduce_off.reduced, key=lambda e: e[0])
    
# http://www.youtube.com/watch?v=Cl3A_9hokjU
g = [None] * 4
g[0] = [0, 1, 0, 1]
g[1] = [1, 0, 1, 1]
g[2] = [0, 1, 0, 1]
g[3] = [1, 1, 1, 0]
assert (3, [0, 1, 0, 2]) == color_vertex(g)

g = [None] * 5
g[0] = [0, 1, 1, 0, 1]
g[1] = [1, 0, 1, 0, 1]
g[2] = [1, 1, 0, 1, 0]
g[3] = [0, 0, 1, 0, 1]
g[4] = [1, 1, 0, 1, 0]
assert (3, [0, 1, 2, 0, 2]) == color_vertex(g)
```

```python
def match1(t, p, any_char):
    return p[0] == any_char or p[0] == t[0]

def fnmatch(t, p): # name, pattern
    if p == "*":
        return True
    if t == "" or p == "":
        return t == "" and p == ""
    if p[0] == "*":
        return fnmatch(t[1:], p) or fnmatch(t, p[1:])
    elif match1(t, p, "?"):
        return fnmatch(t[1:], p[1:])
    else:
        return False

xyz = "xyzz"
assert fnmatch(xyz, "*") 
assert fnmatch("", "") and not fnmatch("x", "") and not fnmatch("", "x")
assert fnmatch(xyz, "*z") and fnmatch(xyz, "x*z") and fnmatch(xyz, "x*")
```

```python
def rematch(t, p):
    if t == "" and p == "":
        return True
    if len(p) > 1 and p[1] == "*":
        return (t and match1(t, p, ".") and rematch(t[1:], p)
               or rematch(t, p[2:]))
    elif t and p and match1(t, p, "."):
        return rematch(t[1:], p[1:])
    else:
        return False
    
assert rematch("abc", ".b.") and rematch("abc", "a.c")
assert not rematch("aaa", "*") and rematch("aaa", "a*")
assert rematch("aaa", "aaa*") and rematch("abc", "a*.*c*")
```

```python
def sum2s(a, b):
    a, b = [list(reversed(e)) for e in (a, b)]
    c = []
    zero = ord("0")
    tens = 0
    for i in range(1 + max(len(a), len(b))):
        ones = tens
        ones += sum(ord(e[i]) - zero for e in (a, b) if i < len(e))
        tens, ones = ones // 10, ones % 10
        c.append(chr(ones + zero))
    return "".join(reversed(c))

assert "1443" == sum2s("456", "987")
```

```python
class DNode:
    def __init__(self, value, pred=None, succ=None):
        self.value, self.pred, self.succ = value, pred, succ

class LRUCache:
    def __init__(self, capacity=1):
        self.capacity = capacity
        self.ha5h = {}
        self.head = self.tail = None

    def put(self, k, v):
        k in self.ha5h and self.delete_node(self.ha5h[k])
        self.push_node(DNode((k, v)))
        self.ha5h[k] = self.tail
        while len(self.ha5h) > self.capacity:
            del self.ha5h[self.shift_node().value[0]]
        return self

    def get(self, k):
        if k in self.ha5h:
            self.delete_node(self.ha5h[k])
            self.push_node(self.ha5h[k])
            return self.tail.value[1]

    def push_node(self, node): # push at tail
        node.succ, node.pred = None, self.tail
        if self.tail:
            self.tail.succ, self.tail = node, node
        else:
            self.head = self.tail = node
        return self

    def delete_node(self, node):
        if self.head != node:
            node.pred.succ = node.succ
        else:
            self.head.succ.pred, self.head = None, self.head.succ
        if self.tail != node:
            node.succ.pred = node.pred
        else:
            self.tail.pred.succ, self.tail = None, self.tail.pred
        return self

    def shift_node(self): # pop at head
        if self.head:
            node = self.head
            if self.head.succ:
                self.head.succ.pred, self.head = None, self.head.succ
            else:
                self.head = self.tail = None
            return node

c = LRUCache(3).put(1, 'a').put(2, 'b').put(3, 'c')
assert 'a' == c.get(1) and 'b' == c.get(2)
assert c.put(4, 'd').get(3) is None
assert c.put(5, 'e').get(1) is None
assert 'b' == c.put(5, 'e').get(2)
```

```python
class CircularBuffer: # http://www.programiz.com/python-programming/exceptions
    def __init__(self, capacity):
        self._a = [None] * (capacity+1)
        self._head = self._tail = 0

    def enq(self, v):
        if self.full():
            raise RuntimeError("This buffer is full.")
        self._a[self._tail] = v
        self._tail = (self._tail+1) % len(self._a)
        return self

    def deq(self):
        if self.empty():
            raise RuntimeError("This buffer is empty.")
        v = self._a[self._head]
        self._head = (self._head+1) % len(self._a)
        return v

    def empty(self):
        return self._head == self._tail

    def full(self):
        return self._head == (self._tail+1) % len(self._a)

buffer = CircularBuffer(3)
assert buffer.empty() and not buffer.full()
buffer.enq(10).enq(20).enq(30)
try:
    buffer.enq(40)
except RuntimeError:
    pass
assert buffer.full()

assert [10, 20, 30] == [buffer.deq() for _ in range(3)]
try:
    buffer.deq()
except RuntimeError:
    pass
assert buffer.empty()
```

Redis Partitioning
* partitioning is splitting data into multiple instances, so each instance only contains a subset of keys.
* two main goals: allowing much larger data sets beyond a memory limit on a single computer, and scaling IO and computational power to multi-cores, computer, and network adapters.
* alternatives
  * range partitioning (user id 0 to 10,000 goes to instance 0).
  * hash partitioning (a hashcode modulo by # of instances): consistent hashing?
  * impl.: client side, proxy assisted, query routing (a rand instance), or a hybrid (redir.).
* disadvantages: intersection between sets, transactions over multiple keys, a big sorted set for a single key, operational complexity, adding or removing capacity.
* re-balancing or presharding in 2^5 instances w/ replication can do the move with minimal or no downtime.
