```python
inf = 2**31 - 1 # or sys.maxvalue

def optimal_tour(g): # TSP http://www.youtube.com/watch?v=aQB_Y9D5pdw
    def popped(s, i):
        l = list(s)
        l.pop(i)
        return tuple(l)
    # g(v, set) is min([g[v, e] + g(e, set-{e}) for e in set])
    def mapped(v, s, memos={}):
        def computed():
            if s:
                return min(
                    ((w, g[v][w] + mapped(w, popped(s, i))[1])
                         for i, w in enumerate(s)
                         if g[v][w] != inf),
                    key=lambda e: e[1])
            else:
                if g[v][0] is inf:
                    return (None, intmax)
                else:
                    return (0, g[v][0])
        return memos.get((v, s)) or memos.setdefault((v, s), computed())
    return mapped(0, tuple(range(1, len(g))))[1]

g = [None] * 4
g[0] = [0, 10, 15, 20]
g[1] = [5, 0, 9, 10]
g[2] = [6, 13, 0, 12]
g[3] = [8, 8, 9, 0]

assert 35 == optimal_tour(g)

def floyd_warshal(g):
    d = [r[:] for r in g]
    n = len(g) # graph in n x n matrix
    for k in range(n):
        for i in range(n):
            for j in range(n):
                d[i][j] = min(d[i][j], d[i][k] + d[k][j])
    return d

floyd_warshal(g)
```

```python
# Given a list of intervals [start, end], collapse intervals that overlap.
def collapsed(intervals):
    intervals = sorted(intervals, key=lambda e: e[0]) # sorted by start.
    collapsed = []
    tail = 0
    for head in range(1, 1 + len(intervals)):
        if head == len(intervals) or intervals[head - 1][1] < intervals[head][0]:
            collapsed.append((intervals[tail][0], intervals[head-1][1]))
            tail = head
    return collapsed

assert [(2, 8), (10, 12)] == collapsed([(2, 6), (4, 8), (10, 12)])
assert [(2, 4), (6, 12)] == collapsed([(2, 4), (6, 10), (8, 12)])
```

```python
# Given a BST of integers, and a range [start, end], find the sum of the values within the range.
def sum_(tree, min_, max_, sum_=None):
    if sum_ is None:
        sum_ = [0]
    if tree is None:
        return 0
    if min_ <= tree.value <= max_:
        sum_[0] += tree.value
    if tree.value >= min_:
        sum_(tree.left, min_, max_, sum_)
    if tree.value <= max_: # 
        sum_(tree.right, min_, max_, sum_)
    return sum_[0]
```

```python
def max_profit(a): # from a list of stock prices.
    max_profit = left = 0
    max_range = (0, 0)
    for right in range(1, len(a)):
        profit = a[right] - a[left]
        if profit < 0:
            left = right
        elif profit > max_profit:
            max_profit, max_range = profit, (left, right)
    return (max_profit, max_range)

assert (30, (2, 3)) == max_profit([30, 40, 20, 50, 10])
assert (0, (0, 0)) == max_profit([30])
```

```python
def minmax(a, start=0, stop=None):
    if stop is None:
        stop = len(a)
    if stop - start < 2:
        return (a[start], a[start])
    else:
        mid = (start + stop - 1)//2
        l = minmax(a, start, mid+1)
        r = minmax(a, mid+1, stop)
        return (min(l[0], r[0]), max(l[1], r[1]))

assert (1, 9) == minmax([1, 3, 5, 7, 9, 2, 4, 6, 8])

def peak(a, start=0, stop=None):
    if stop is None:
        stop = len(a)
    mid = (start + stop - 1)//2
    if a[mid-1] < a[mid] > a[mid+1]:
        return mid
    elif a[mid-1] < a[mid] < a[mid+1]:
        return peak(a, mid+1, stop)
    else:
        return peak(a, start, mid)

assert 6 == peak([1, 5, 7, 9, 15, 18, 21, 19, 14])
```

```python
def three_sum_closest(a, q=0):
    a, n = sorted(a), len(a)
    tuples = []
    for i in range(n-2):
        j, k = i+1, n-1
        while j < k:
            diff = abs(q - sum((a[i], a[j], a[k])))
            if not tuples or mindiff > diff:
                mindiff = diff
                tuples = [(a[i], a[j], a[k])]
            elif mindiff == diff:
                tuples.append((a[i], a[j], a[k]))
            cmp = q - sum((a[i], a[j], a[k]))
            if cmp < 0:
                j += 1
            elif cmp > 0:
                k -= 1
            else:
                j += 1
                k -= 1
    return tuples

def three_sum(a, q=0):
    a, n = sorted(a), len(a)
    tuples = []
    for i in range(n-3):
        j, k = i+1, n-1
        while j < k:
            cmp = q - sum((a[i], a[j], a[k]))
            if cmp < 0:
                j += 1
            elif cmp > 0:
                k -= 1
            else:
                tuples.append((a[i], a[j], a[k]))
                j += 1
                k -= 1
    return tuples

assert [(-1, 1, 2)] == three_sum_closest([-1, 2, 1, -4])
assert [(-1, -1, 2), (-1, 0, 1)] == three_sum([-1, 0, 1, 2, -1, -4], 0)
```

```python
def next_permutate(a, k=None):
    n, k = len(a), k or 1
    for _ in range(k):
        i = next((i for i in range(n-2, -1, -1) if a[i] < a[i+1]), None)
        if i is None:
            return None
        j = next(j for j in range(n-1, i, -1) if a[i] < a[j])
        a[i], a[j] = a[j], a[i]
        a[i+1:] = a[-1:i:-1]
    return a[:]

a = [1, 2, 3, 4]
p = [[1, 2, 4, 3], [1, 3, 2, 4], [4, 3, 2, 1], None]
assert p == [next_permutate(a, k) for k in (1, 1, 21, 1)]
```

```python
def diameter(tree): # return (d, h)
    if tree is None:
        return (0, 0)
    l = diameter(tree.left)
    r = diameter(tree.right)
    d = max(l[0], r[0], 1 + l[1] + r[1])
    h = 1 + max(l[1], r[1])
    return (d, h)

# tree input:   a
#             b
#          c    f
#           d     g
#            e
tree = parse('abcdefg', 'cdebfga')
assert (6, 5) == diameter(tree)
```

```python
def backtrack(candidate, expand_out, reduce_off):
    if not reduce_off(candidate):
        for e in expand_out(candidate):
            candidate.append(e)
            backtrack(candidate, expand_out, reduce_off)
            candidate.pop()

def color_vertex(g):
    def expand_out(a):
        v = len(a) # vertex v
        return [c for c in set(a) if peaceful_at(a, v, c)] + [max(a)+1]

    def peaceful_at(a, v, c):
        return all([g[v][w] == 0 or a[w] != c for w in range(v)])

    def reduce_off(a):
        if reduce_off.reduced or len(a) == len(g):
            reduce_off.reduced.append((max(a)+1, a[:]))
            return True

    reduce_off.reduced = []
    backtrack([0], expand_out, reduce_off)
    return min(reduce_off.reduced, key=lambda e: e[0])
    
# http://www.youtube.com/watch?v=Cl3A_9hokjU
g = [None] * 4
g[0] = [0, 1, 0, 1]
g[1] = [1, 0, 1, 1]
g[2] = [0, 1, 0, 1]
g[3] = [1, 1, 1, 0]
assert (3, [0, 1, 0, 2]) == color_vertex(g)

g = [None] * 5
g[0] = [0, 1, 1, 0, 1]
g[1] = [1, 0, 1, 0, 1]
g[2] = [1, 1, 0, 1, 0]
g[3] = [0, 0, 1, 0, 1]
g[4] = [1, 1, 0, 1, 0]
assert (3, [0, 1, 2, 0, 2]) == color_vertex(g)
```

```python
def match1(t, p, any_char):
    return p[0] == any_char or p[0] == t[0]

def fnmatch(t, p): # name, pattern
    if p == "*":
        return True
    if t == "" or p == "":
        return t == "" and p == ""
    if p[0] == "*":
        return fnmatch(t[1:], p) or fnmatch(t, p[1:])
    elif match1(t, p, "?"):
        return fnmatch(t[1:], p[1:])
    else:
        return False

xyz = "xyzz"
assert fnmatch(xyz, "*") 
assert fnmatch("", "") and not fnmatch("x", "") and not fnmatch("", "x")
assert fnmatch(xyz, "*z") and fnmatch(xyz, "x*z") and fnmatch(xyz, "x*")
```

```python
def rematch(t, p):
    if t == "" and p == "":
        return True
    if len(p) > 1 and p[1] == "*":
        return (t and match1(t, p, ".") and rematch(t[1:], p)
               or rematch(t, p[2:]))
    elif t and p and match1(t, p, "."):
        return rematch(t[1:], p[1:])
    else:
        return False
    
assert rematch("abc", ".b.") and rematch("abc", "a.c")
assert not rematch("aaa", "*") and rematch("aaa", "a*")
assert rematch("aaa", "aaa*") and rematch("abc", "a*.*c*")
```

```python
def sum2s(a, b):
    a, b = [list(reversed(e)) for e in (a, b)]
    c = []
    zero = ord("0")
    tens = 0
    for i in range(1 + max(len(a), len(b))):
        ones = tens
        ones += sum(ord(e[i]) - zero for e in (a, b) if i < len(e))
        tens, ones = ones // 10, ones % 10
        c.append(chr(ones + zero))
    return "".join(reversed(c))

assert "1443" == sum2s("456", "987")
```

```python
class DNode:
    def __init__(self, value, pred=None, succ=None):
        self.value, self.pred, self.succ = value, pred, succ

class LRUCache:
    def __init__(self, capacity=1):
        self.capacity = capacity
        self.ha5h = {}
        self.head = self.tail = None

    def put(self, k, v):
        k in self.ha5h and self.delete_node(self.ha5h[k])
        self.push_node(DNode((k, v)))
        self.ha5h[k] = self.tail
        while len(self.ha5h) > self.capacity:
            del self.ha5h[self.shift_node().value[0]]
        return self

    def get(self, k):
        if k in self.ha5h:
            self.delete_node(self.ha5h[k])
            self.push_node(self.ha5h[k])
            return self.tail.value[1]

    def push_node(self, node): # push at tail
        node.succ, node.pred = None, self.tail
        if self.tail:
            self.tail.succ, self.tail = node, node
        else:
            self.head = self.tail = node
        return self

    def delete_node(self, node):
        if self.head != node:
            node.pred.succ = node.succ
        else:
            self.head.succ.pred, self.head = None, self.head.succ
        if self.tail != node:
            node.succ.pred = node.pred
        else:
            self.tail.pred.succ, self.tail = None, self.tail.pred
        return self

    def shift_node(self): # pop at head
        if self.head:
            node = self.head
            if self.head.succ:
                self.head.succ.pred, self.head = None, self.head.succ
            else:
                self.head = self.tail = None
            return node

c = LRUCache(3).put(1, 'a').put(2, 'b').put(3, 'c')
assert 'a' == c.get(1) and 'b' == c.get(2)
assert c.put(4, 'd').get(3) is None
assert c.put(5, 'e').get(1) is None
assert 'b' == c.put(5, 'e').get(2)
```

```python
class CircularBuffer: # http://www.programiz.com/python-programming/exceptions
    def __init__(self, capacity):
        self._a = [None] * (capacity+1)
        self._head = self._tail = 0

    def enq(self, v):
        if self.full():
            raise RuntimeError("This buffer is full.")
        self._a[self._tail] = v
        self._tail = (self._tail+1) % len(self._a)
        return self

    def deq(self):
        if self.empty():
            raise RuntimeError("This buffer is empty.")
        v = self._a[self._head]
        self._head = (self._head+1) % len(self._a)
        return v

    def empty(self):
        return self._head == self._tail

    def full(self):
        return self._head == (self._tail+1) % len(self._a)

buffer = CircularBuffer(3)
assert buffer.empty() and not buffer.full()
buffer.enq(10).enq(20).enq(30)
try:
    buffer.enq(40)
except RuntimeError:
    pass
assert buffer.full()

assert [10, 20, 30] == [buffer.deq() for _ in range(3)]
try:
    buffer.deq()
except RuntimeError:
    pass
assert buffer.empty()
```

Given two strings of size m, n and operations of insert (I), delete (D), and replace (R)
all at equal cost. Find the minimum # of edits required to transform a string into another.

```python
from enum import Enum
Edit = Edit = Enum("Edit", {"I": "Insert", "D": "Delete", "R": "Replace", "M": "Match"})

def edit(a, b, match=lambda c1, c2: int(c1 != c2)):
    memos = {}
    def mapped(i, o):
        def compute(i, o):
            if i == 0 or o == 0:
                return (None, i + o)
            else:
                return min((
                    (Edit.M, mapped(i-1, o-1)[1] + match(a[i-1], b[o-1])),
                    (Edit.D, mapped(i-1, o)[1] + 1),
                    (Edit.I, mapped(i, o-1)[1] + 1)
                ), key=lambda e: e[1])
        if (i, o) in memos:
            return memos[(i, o)]
        else:
            return memos.setdefault((i, o), compute(i, o))

    def reduced(i, o):
        m = memos[(i, o)]
        if m[0] is None:
            return ()
        if m[0] is Edit.M:
            return reduced(i-1, o-1) + (Edit.M if a[i-1] == b[o-1] else Edit.R, )
        if m[0] is Edit.I:
            return reduced(i, o-1) + (Edit.I,)
        if m[0] is Edit.D:
            return reduced(i-1, o) + (Edit.D,)

    m, n = len(a), len(b)
    distance = mapped(m, n)[1]
    return (distance, reduced(m, n))

assert (3, (Edit.R, Edit.M, Edit.M, Edit.M, Edit.R, Edit.M, Edit.I)) == edit('kitten', 'sitting')
```

    private static final int N_LOCKS = 16;
    private final Node[] buckets;
    private final Object[] locks;

    private static class Node {
        Object key;
        Object value;
        Node next;
    }

    public StripedMap(int buckets) {
        this.buckets = new Node[buckets];
        this.locks = new Object[N_LOCKS];
        for (int i = 0; i < N_LOCKS; i++) {
            this.locks[i] = new Object();
        }
    }

    public Object get(Object key) {
        int hash = hash(key);
        synchronized (locks[hash % N_LOCKS]) {
            for (Node m = buckets[hash]; m != null; m = m.next) {
                if (m.key.equals(key)) {
                    return m.value;
                }
            }
        }
        return null;
    }

    public void clear() {
        for (int i = 0; i < buckets.length; i++) {
            synchronized (locks[i % N_LOCKS]) {
                buckets[i] = null;
            }
        }
    }

    private int hash(Object key) {
        return Math.abs(key.hashCode() % buckets.length);

#### Listenable Future explained

There are simple, but powerful abstractions, such as **[FutureTask](http://docs.oracle.com/javase/7/docs/api/java/util/concurrent/FutureTask.html)** (a result-bearing computation) with methods: run(), cancel(interrupt\_if\_running), get(timeout), is\_cancelled(), and is\_done(), and **[ListenableFuture](https://code.google.com/p/guava-libraries/wiki/ListenableFutureExplained)** with a method: **add\_listener(runnable\_or\_callback)** when the future task is complete, where the listener implements **on_complete()**, or **on\_success()** & **on_failure()**. callback\_on\_all, callback\_on\_any.
