```python
# [SortedDict](http://www.grantjenks.com/docs/sortedcontainers/sorteddict.html)
# * pop(key, default), popitem()
# * index(key, start, stop), iloc[pos]
# * bisect_left(key), bisect_right(key)
from sortedcontainers import SortedDict # pip3 install sortedcontainers
from collections import deque

def closest_pairs(points):
    sq_dist = lambda p1, p2: sum((p1[i] - p2[i])**2 for i in range(2))
    m = SortedDict()
    q = deque(sorted(points, key=lambda e: e[0]))
    cp = None # closest pairs
    p0 = q.popleft()
    m[p0[1]] = set([p0])
    while q:
        p1 = q.popleft()
        if not cp:
            cp, d = [(p0, p1)], sq_dist(p0, p1)
        else:
            m.bisect_left(d)
            # do some
        if p1[1] not in m:
            m[p1[1]] = set()
        m[p1[1]].add(p1)
    return points

points = ((-2, -2), (4, -8), (8, -2), (10, 4), (13, -8), (14, 0), (17, -6))

m = SortedDict()
for e in range(10):
    m[10 * e] = e
print(m)
# closest_pairs(points)
```

```python
def dot(p1, p2, p3): # a positive means -90' to 90'; 0 means perpendicular.
    x1, y1, x2, y2, x3, y3 = [e for p in (p1, p2, p3) for e in p]
    return (x2 - x1) * (x3 - x1) + (y2 - y1) * (y3 - y1)
    
def cross(p1, p2, p3): # a positve means clockwise; 0 means collinear.
    x1, y1, x2, y2, x3, y3 = [e for p in (p1, p2, p3) for e in p]
    return (x2 - x1) * (y3 - y1) -  (x3 - x1) * (y2 - y1)

def in_range(p, rect): # a point is in a rect of a line.
    px, py = p[0], p[1]
    xx, yy = [sorted(e[i] for e in rect) for i in range(2)]
    return all((xx[0] <= px <= xx[-1], yy[0] <= py <= yy[-1]))

def intersecting(ln1, ln2):
    p1, q1, p2, q2 = [e for e in (ln1 + ln2)]
    c1, c2, c3, c4 = [cross(*e) for e in (
        (p1, q1, p2), (p1, q1, q2), (p2, q2, p1), (p2, q2, q1))]
    return any((
        (c1 * c2 < 0 and c3 * c4 < 0),
        (c1 == 0 and in_range(p2, ln1)),
        (c2 == 0 and in_range(q2, ln1)),
        (c3 == 0 and in_range(p1, ln2)),
        (c4 == 0 and in_range(q1, ln2))))

p, q, r, s, t = ((4, 1), (1, 4), (1, 1), (-3, -3), (4, 4))
c_cases = ((p, q, r), (q, r, p), (p, r, q), (r, q, p), (p, s, t))
assert [9, 9, -9, -9, -21] == [cross(*e) for e in c_cases]

d_cases = ((p, q, t), (t, p, q), (r, s, t))
assert [9, 0, -24] == [dot(*e) for e in d_cases]

e_cases = ((p, q, r, s), (p, q, s, t))
assert [0, 1] == [intersecting(e[:2], e[2:]) for e in e_cases]
```

```python
def is_inside(p, polygon):
    n = len(polygon)
    if n < 3:
        return False
    p2 = (2 * 31 - 1, p[1])
    intersections = 0
    for i in range(n):
        p3, p4 = polygon[i], polygon[(i+1)%n]
        if cross(p, p3, p4) == 0 and in_range(p, (p3, p4)):
            return True
        if intersecting((p, p2), (p3, p4)):
            intersections += 1
    return intersections % 2 == 1

polygons = (
    ((0, 0), (10, 0), (10, 10), (0, 10)),
    ((0, 0), (5, 5), (5, 0)))
points = ((13, 13), (7, 7), (3, 3), (4, 1))
assert ([0, 1, 1, 1, 0, 0, 1, 1] == 
    [int(is_inside(pt, pl)) for pl in polygons for pt in points])
```

```python
def area(polygon):
    n = len(polygon)
    return abs(sum(cross(polygon[0], polygon[i], polygon[i+1])
                   for i in range(1, n-1)) / 2.0)

polygon = ((0, 0), (4, 0), (4, 4), (0, 4), (2, 2))
assert 12 == area(polygon) and 12 == area(list(reversed(polygon)))
```

```python
from math import *

def p2p_dist(p1, p2):
    x1, y1, x2, y2 = [e for p in (p1, p2) for e in p]
    return sqrt((x2-x1)**2 + (y2-y1)**2)

def l2p_dist(p, ln, segmented=False):
    dist = cross(p, ln[0], ln[1]) / p2p_dist(ln[0], ln[1])
    if segmented:
        if dot(ln[1], ln[0], p) > 0:
            return p2p_dist(ln[0], p)
        if dot(ln[0], ln[1], p) > 0:
            return p2p_dist(ln[1], p)
    return abs(dist);

assert 5.0 == l2p_dist((-3, -2), ((1, 1), (5, 5)), True)
```
