```python
def angle_sign(p1, p2, p3): # http://en.wikipedia.org/wiki/Cross_product#Computational_geometry
    x1, y1, x2, y2, x3, y3 = [e for p in (p1, p2, p3) for e in p]
    angle = (x2 - x1) * (y3 - y1) - (y2 - y1) * (x3 - x1)
    return angle

def is_online(p, ln):
    px, py = p[0], p[1]
    xx, yy = [sorted(e[i] for e in ln) for i in range(2)]
    return all((xx[0] <= px <= xx[-1], yy[0] <= py <= yy[-1]))

def intersecting(ln1, ln2):
    p1, q1, p2, q2 = [e for e in (ln1 + ln2)]
    s1, s2, s3, s4 = [angle_sign(*e) for e in (
        (p1, q1, p2), (p1, q1, q2), (p2, q2, p1), (p2, q2, q1))]
    return any((
        (s1 * s2 < 0 and s3 * s4 < 0),
        (s1 == 0 and is_online(p2, ln1)),
        (s2 == 0 and is_online(q2, ln1)),
        (s3 == 0 and is_online(p1, ln2)),
        (s4 == 0 and is_online(q1, ln2))))

p, q, r, s, t = ((1, 1), (4, 2), (2, 4), (4, 4), (2, 2))
d_cases = ((p, q, r), (q, r, p), (r, p, q), (p, r, q), (r, q, p), (q, p, r), (p, s, t))
assert [8, 8, 8, -8, -8, -8, 0] == [angle_sign(*e) for e in d_cases]

e_cases = ((p, q, r, s), (q, r, s, t), (p, r, q, s), (p, s, t, q))
assert [0, 1, 0, 1] == [intersecting(e[:2], e[2:]) for e in e_cases]
```

```python
# http://www.geeksforgeeks.org/how-to-check-if-a-given-point-lies-inside-a-polygon/
def is_inside(p, polygon):
    n = len(polygon)
    if n < 3:
        return False
    p2 = (2 * 31 - 1, p[1])
    intersections = 0
    for i in range(n):
        p3, p4 = polygon[i], polygon[(i+1)%n]
        if angle_sign(p, p3, p4) == 0 and is_online(p, (p3, p4)):
            return True
        if intersecting((p, p2), (p3, p4)):
            intersections += 1
    return intersections % 2 == 1

polygons = (
    ((0, 0), (10, 0), (10, 10), (0, 10)),
    ((0, 0), (5, 5), (5, 0)))
points = ((13, 13), (7, 7), (3, 3), (4, 1))
assert ([0, 1, 1, 1, 0, 0, 1, 1] == 
    [int(is_inside(pt, pl)) for pl in polygons for pt in points])
```

```python
# http://www.grantjenks.com/docs/sortedcontainers/sorteddict.html
from sortedcontainers import SortedDict
import heapq

def intersected(line_segs): # find intersecting line segments.
    line_segs = sorted(line_segs, key=lambda e: e[0])
    t = SortedDict()
    q = []
    res = []
    for ln in line_segs:
        x1, y1, x2, y2 = [e for p in ln for e in p]
        while q and x1 > q[0][0]:
            ln2 = heapq.heappop(q)[1]
            ln2y1 = ln2[0][1]
            t[ln2y1].remove(ln2)
        if t:
            r = t.bisect_right(y1)
            if r > 0: # below
                res.extend((ln, ln2) for ln2 in t.values()[r-1] if intersecting(ln, ln2))
            if r < len(t): # above
                res.extend((ln, ln2) for ln2 in t.values()[r] if intersecting(ln, ln2))
        heapq.heappush(q, (x2, ln))
        if y2 not in t:
            t[y2] = set()
        t[y2].add(ln)
    return res

assert ([(((2, 4), (6, 0)), ((1, 1), (4, 4))), (((3, 4), (7, 0)), ((1, 1), (4, 4)))]
        == intersected((((1, 1), (4, 4)), ((2, 4), (6, 0)), ((3, 4), (7, 0)))))
```
